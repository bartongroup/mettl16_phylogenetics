```python
import re
import itertools as it
from operator import itemgetter
import subprocess as sp
import json

import numpy as np
import pandas as pd
import networkx as nx

from Bio import PDB as pdb


def domain_pos_to_invs(dom_pos):
    dom_pos = np.array(sorted(list(dom_pos)))
    d_idx = np.flatnonzero(np.diff(dom_pos) > 1)
    left_idx = 0
    invs = []
    for idx in d_idx:
        invs.append([dom_pos[left_idx], dom_pos[idx]])
        left_idx = idx + 1
    else:
        invs.append([dom_pos[left_idx], dom_pos[-1]])
    return invs


def find_domains(plddt, pae, min_plddt=70, min_pae=5, min_res=20, pae_expo=-1):
    g = nx.Graph()
    for i, j in it.combinations(range(len(plddt)), r=2):
        if plddt[i] < min_plddt or plddt[j] < min_plddt:
            continue
        else:
            p_ij = min(pae[i, j], pae[j, i])
            if p_ij <= min_pae:
                g.add_edge(i, j, weight=float(p_ij) ** pae_expo)
    domains = []
    for clust in nx.algorithms.community.greedy_modularity_communities(g, weight='weight'):
        if len(clust) >= min_res:
            domains.append([np.median(list(clust)), len(clust), clust])
    domains.sort(key=itemgetter(0))
    return domains


def get_plddt(struct):
    return np.array(
        [aa['CA'].get_bfactor() for aa in struct.get_residues()]
    )


def get_nres(pdb_fn):
    struct = pdb.PDBParser().get_structure(pdb_fn, pdb_fn)
    return len(list(struct.get_residues()))


def get_pae(pae_fn):
    with open(pae_fn) as f:
        return np.array(json.load(f)[0]['predicted_aligned_error'])


class HighConfDomainSelect(pdb.Select):

    def __init__(self, domain_resi, min_b_factor):
        super().__init__()
        self.domain_resi = domain_resi
        self.min_b_factor = min_b_factor
    
    def accept_residue(self, residue):
        if residue['CA'].get_bfactor() >= self.min_b_factor:
            _, pos, _ = residue.get_id()
            if pos in self.domain_resi:
                return 1
            else:
                return 0
        else:
            return 0


def write_domain(struct, output_fn, domain_resi, min_plddt=70):
    io = pdb.PDBIO()
    io.set_structure(struct)
    selecter = HighConfDomainSelect(domain_resi, min_plddt)
    io.save(output_fn, selecter)


def generate_domain_pdbs(pdb_fn, pae_fn, species_name, basename, output_dir, min_plddt=70, min_pae=5, min_res=20):
    struct = pdb.PDBParser().get_structure(species_name, pdb_fn)
    plddt = get_plddt(struct)
    pae = get_pae(pae_fn)
    domains = find_domains(plddt, pae, min_plddt, min_pae, min_res)

    domain_data = []
    for i, (med_pos, nres, resi) in enumerate(domains, 1):
        dom_name = f'{basename}_dom{i}of{len(domains)}'
        highconf_fn = os.path.join(output_dir, f'{dom_name}.highconf.pdb')
        full_fn = os.path.join(output_dir, f'{dom_name}.full.pdb')
        dom_invs = domain_pos_to_invs(resi)
        dom_start, dom_end = dom_invs[0][0], dom_invs[-1][1]
        write_domain(struct, highconf_fn, resi, min_plddt)
        write_domain(struct, full_fn, np.arange(dom_start, dom_end + 1), 0)
        domain_data.append([species_name, pdb_fn, dom_name, med_pos, nres, dom_start, dom_end, dom_invs, highconf_fn, full_fn])
    return pd.DataFrame(
        domain_data,
        columns=['species', 'pdb_fn', 'dom_name', 'median_position', 'dom_nres', 'dom_start', 'dom_end', 'dom_invs', 'dom_fn_highconf', 'dom_fn_full']
    )
```

```python
def get_file_basename(fn):
    return os.path.splitext(os.path.split(fn)[1])[0]

species = {v: k  for k, v in snakemake.config['comps'][snakemake.wildcards.comp].items()}
pdb_fns = {get_file_basename(fn): fn for fn in snakemake.input.af_pdb}
pae_fns = {get_file_basename(fn): fn for fn in snakemake.input.af_pae}


doms = []
for basename, pdb_fn in pdb_fns.items():
    pae_fn = pae_fns[basename]
    sp = species[basename]
    doms.append(
        generate_domain_pdbs(
            pdb_fn, pae_fn, sp,
            f'{snakemake.wildcards.comp}_{basename}',
            'domains'
        )
    )

doms = pd.concat(doms)
doms.head()
```

```python
import os
import shutil
import re
from glob import glob
import itertools as it
import tempfile
import warnings
import subprocess as sp

import numpy as np
import pandas as pd
from scipy import stats
from scipy.cluster.hierarchy import linkage, optimal_leaf_ordering, leaves_list
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt
from matplotlib import patches
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import seaborn as sns

import ete3
from Bio import PDB as pdb
os.environ['QT_QPA_PLATFORM']='offscreen'

if shutil.which('../scripts/TMalign') is None:
    raise RuntimeError('Place TMalign binary in scripts dir')
if shutil.which('../scripts/USalign') is None:
    raise RuntimeError('Place USalign binary in scripts dir')

## Default plotting params
%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
def get_nres(pdb_fn):
    struct = pdb.PDBParser().get_structure(pdb_fn, pdb_fn)
    return len(list(struct.get_residues()))


def get_file_basename(fn):
    return os.path.splitext(os.path.split(fn)[1])[0]


doms = pd.read_csv(snakemake.input.doms_csv, index_col=0)
ref_struct_fns = {get_file_basename(fn): fn for fn in snakemake.input.ref_doms}
ref_doms = pd.DataFrame([
    ['Homo_sapiens', None, r, None, get_nres(fn), None, None, None, fn, fn]
    for r, fn in ref_struct_fns.items()
], columns=doms.columns)
doms = pd.concat([doms, ref_doms])
doms.tail()
```

```python
def calculate_tmscore(pdb1_fn, pdb2_fn):
    res = sp.run(
        ['../scripts/TMalign', pdb1_fn, pdb2_fn],
        stdout=sp.PIPE
    ).stdout.decode().split('\n')
    tm1 = re.search('^TM-score= ([10]\.\d+) ', res[13]).group(1)
    tm2 = re.search('^TM-score= ([10]\.\d+) ', res[14]).group(1)
    
    return max(float(tm1), float(tm2))


def pairwise_tmscores(doms):
    tmscore = np.empty(shape=(len(doms), len(doms)))

    for i, dom1 in enumerate(doms.itertuples(index=False)):
        for j, dom2 in enumerate(doms.itertuples(index=False)):
            if dom1.dom_name == dom2.dom_name:
                tmscore[i, j] = 1
            else:
                tm = calculate_tmscore(dom1.dom_fn_highconf, dom2.dom_fn_highconf)
                tmscore[i, j] = tm

    tmscore = pd.DataFrame(tmscore, columns=doms.dom_name, index=doms.dom_name)
    return tmscore


tmscores = pairwise_tmscores(doms)
tmscores.head()
```

```python
def generate_clusters(cluster1_name, cluster1_seed_name,
                      cluster2_name, cluster2_seed_name,
                      tmscores, threshold=0.45):

    both_name = f'{cluster1_name}_{cluster2_name}'
    clusters = {
        cluster1_name: set(),
        both_name: set(),
        cluster2_name: set(),
        'other': set()
    }

    for n, row in tmscores.iterrows():
        c1 = row.loc[cluster1_seed_name] > threshold
        c2 = row.loc[cluster2_seed_name] > threshold
        if c1 and c2:
            clusters[both_name].add(n)
        elif c1:
            clusters[cluster1_name].add(n)
        elif c2:
            clusters[cluster2_name].add(n)
        else:
            clusters['other'].add(n)


    ordered_clusters = {}

    for name, clust in clusters.items():
        labels = list(clust)
        if len(labels):
            dist = squareform(np.abs(np.log(tmscores.loc[labels, labels])))
            idx = leaves_list(optimal_leaf_ordering(linkage(dist), dist))
            ordered_clusters[name] = [
                labels[i] for i in idx
            ]
        else:
            ordered_clusters[name] = []

    order = ordered_clusters[cluster1_name] + ordered_clusters[both_name] + ordered_clusters[cluster2_name]

    return ordered_clusters, order


clusters, order = generate_clusters(
    'mtd', snakemake.config['reference_xray_structures']['mtd'],
    'vcr', snakemake.config['reference_xray_structures']['vcr'],
    tmscores, threshold=0.45
)
```

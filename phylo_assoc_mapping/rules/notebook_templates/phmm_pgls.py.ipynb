```python
import os
import re
from glob import glob
import tempfile
import warnings
import itertools as it
import tempfile

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap, to_hex
import seaborn as sns

from pgls_tools import pssm_phmm_pgls
from pgls_tools.og import read_orthofinder_n0
from pgls_tools.phmm import read_fasta

## Default plotting params
%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
IUPAC = {
    'A': ['A'],
    'C': ['C'],
    'G': ['G'],
    'U': ['U'],
    'R': ['A', 'G'],
    'Y': ['C', 'U'],
    'S': ['G', 'C'],
    'W': ['A', 'U'],
    'K': ['G', 'U'],
    'M': ['A', 'C'],
    'B': ['C', 'G', 'U'],
    'D': ['A', 'G', 'H'],
    'H': ['A', 'C', 'U'],
    'V': ['A', 'C', 'G'],
    'N': ['A', 'C', 'G', 'U']
}


sign, pos, nom_base, denom_base = re.search(
    '([pm])(\d+)([ACGURYSKMWBDHVN])to([ACGURYSKMWBDHVN])',
    snakemake.wildcards.phenotype
).groups()

abspos = int(pos) - 1 if sign == 'p' else -int(pos)
if snakemake.wildcards.signal_type == '5ss':
    abspos = 3 + abspos
print(abspos)

nom_bases, denom_bases = IUPAC[nom_base], IUPAC[denom_base]
print(nom_bases, denom_bases)
```

```python
og, og_cnv = read_orthofinder_n0(
    snakemake.input.key_ogs,
    snakemake.input.key_prot_ids,
)
og_prot_ids = og.loc[snakemake.wildcards.og_id].to_dict()
```

```python
og_seqs = {}
for fasta_fn in snakemake.input.prots:
    species = os.path.splitext(os.path.split(fasta_fn)[1])[0]
    prot_ids = og_prot_ids[species]
    if len(prot_ids):
        prot_ids = prot_ids.split(', ')
        seqs = read_fasta(fasta_fn)
        og_seqs[species] = [seqs[pid] for pid in prot_ids]
```

```python
h, fasta_fn = tempfile.mkstemp(dir=os.environ['TMPDIR'])
with open(fasta_fn, 'w') as fasta:
    for sp, seqs in og_seqs.items():
        for seq in seqs:
            fasta.write(f'>{sp}\n{seq}\n')

res = pssm_phmm_pgls(
    fasta_fn,
    snakemake.input.phmm,
    snakemake.input.pssms,
    snakemake.input.tree,
    abspos, nom_bases, denom_bases,
    njobs=8,#snakemake.threads
)

os.close(h)
os.remove(fasta_fn)
```

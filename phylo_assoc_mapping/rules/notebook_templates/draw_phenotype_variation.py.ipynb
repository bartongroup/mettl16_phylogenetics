```python
import os
import re
from glob import glob
import tempfile
import warnings

import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap, to_hex
import seaborn as sns
import ete3

## Default plotting params
%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
pssms = {
    os.path.split(fn)[1].split('.')[0]: pd.read_csv(fn, sep='\t', index_col=[0, 1])
    for fn in snakemake.input.pssms
}
```

```python
import itertools as it
from collections import defaultdict

comps = list(it.chain(
    it.combinations('ACGU', r=2),
    zip('ACGUSR', 'BDHVWY')
))


IUPAC = {
    'A': ['A'],
    'C': ['C'],
    'G': ['G'],
    'U': ['U'],
    'R': ['A', 'G'],
    'Y': ['C', 'U'],
    'S': ['G', 'C'],
    'W': ['A', 'U'],
    'K': ['G', 'U'],
    'M': ['A', 'C'],
    'B': ['C', 'G', 'U'],
    'D': ['A', 'G', 'U'],
    'H': ['A', 'C', 'U'],
    'V': ['A', 'C', 'G'],
    'N': ['A', 'C', 'G', 'U']
}


def relstd(arr):
    return np.std(arr) / np.abs(np.mean(arr)) * 100


mean_pssms = [pssm.groupby(level='position').mean() for pssm in pssms.values()]

pheno = defaultdict(list)
for pos in range(10):
    for nom_base, denom_base in comps:
        p = []
        for pssm in mean_pssms:
            n = pssm.loc[pos, IUPAC[nom_base]].sum()
            d = pssm.loc[pos, IUPAC[denom_base]].sum()
            p.append(n - d)
        pheno[(nom_base, denom_base)].append(np.std(p))

pheno = pd.DataFrame.from_dict(pheno, orient='index')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

HIGHLIGHT_COMPS = {
    ('A', 'C'): pal[5],
    ('A', 'G'): pal[1],
    ('A', 'U'): pal[0],
    ('C', 'G'): pal[3],
    ('C', 'U'): pal[4],
    ('G', 'U'): pal[2]
}


for comp in HIGHLIGHT_COMPS:
    row = pheno.loc[[comp]]
    ax.plot(
        np.arange(10),
        row.values.ravel(),
        lw=2,
        color=HIGHLIGHT_COMPS[comp],
        label=f'{comp[0]} to {comp[1]} ratio',
        zorder=2
    )

ax.legend(title='Phenotype')
ax.set_ylabel('Variation in phenotype across\nspecies (standard deviation)')
ax.set_xticks(np.arange(10))
ax.set_xticklabels(['–3', '–2', '–1', '+1', '+2', '+3', '+4', '+5', '+6', '+7'])
ax.set_xlabel('5\'SS position')
ax.set_ylim(-0.02, 0.55)
plt.savefig(snakemake.output.pairwise_plot)
plt.show()
```

```python
def to_ape_tree(tree):
    tmp_fh, tmp_fn = tempfile.mkstemp()
    tree.write(outfile=tmp_fn, format=1, format_root_node=True)
    r_tree = ape.read_tree(tmp_fn)
    os.close(tmp_fh)
    os.remove(tmp_fn)
    return r_tree


def get_ancestral_states(tree, pheno):
    r_tree = to_ape_tree(tree)
    leaf_order = list(r_tree.rx2['tip.label'])
    pheno = [pheno[n] for n in leaf_order]
    anc = ape.ace(
        robj.FloatVector(pheno),
        r_tree,
        type='continuous'
    )
    return dict(zip(r_tree.rx2['node.label'], anc.rx2['ace']))


def _plot_recursive(node, pheno, x1, y1, is_loss, loss_nodes, ax):
    if not is_loss:
        is_loss = node.name in loss_nodes
    colour = pal[1] if is_loss else pal[0]
    zorder = 1 if is_loss else 0
    for child in node.children:
        x2 = x1 + child.dist
        y2 = pheno[child.name]
        node_leaves = set([n.name for n in child.get_leaves()])
        lw = 1.5
        alpha = 0.75
        ax.plot([x1, x2], [y1, y2], color=colour, lw=lw, zorder=zorder, alpha=alpha)
        if not child.is_leaf():
            _plot_recursive(child, pheno, x2, y2, is_loss, loss_nodes, ax)


def plot_stacked_pheno_histogram(pheno, og_pav, ax):
    hist_data = []
    colours = []
    labels = []
    for pav_type, species_list in og_pav.groupby(og_pav):
        species_list = species_list.index.tolist()
        hist_data.append([pheno[sp] for sp in species_list])
        colours.append({True: pal[0], False: pal[1]}[pav_type])
        labels.append({True: 'Present', False: 'Absent'}[pav_type])

    for i in range(len(hist_data)):
        ax.hist(
            np.concatenate(hist_data[i:]),
            orientation='horizontal',
            range=(-1, 1),
            bins=20,
            color=colours[i],
            label=labels[i],
            edgecolor='none',
        )
    ax.legend(title=og_pav.name)
    return ax



def plot_phenogram(tree, pheno, og_pav,
                   title=None, ultrametric=True,
                   figsize=(10, 5), add_hist=True):
    tree = tree.copy()
    if ultrametric:
        tree.convert_to_ultrametric(tree_length=1)
    anc_states = get_ancestral_states(tree, pheno)
    anc_states.update(pheno)

    fig = plt.figure(figsize=figsize)
    if add_hist:
        phenogram_ax = plt.subplot2grid((1, 4), (0, 0), colspan=2)
        hist_ax = plt.subplot2grid((1, 4), (0, 3), colspan=1, sharey=phenogram_ax)
        plot_stacked_pheno_histogram(pheno, og_pav, hist_ax)
        hist_ax.set_xlabel('Species')
    else:
        phenogram_ax = plt.subplots()
        hist_ax = None
    x1 = 0
    y2 = anc_states[tree.name]
    loss_nodes = pgls_tree.count_losses(tree, og_pav)
    is_loss = tree.name in loss_nodes
    _plot_recursive(tree, anc_states, x1, y2, is_loss, loss_nodes, phenogram_ax)
    if ultrametric:
        phenogram_ax.set_xlabel('Branch length (ultrametric)')
    else:
        phenogram_ax.set_xlabel('Branch length')
    if title is not None:
        phenogram_ax.set_ylabel(title)
    return fig, phenogram_ax, hist_ax
```

```python
pos_name = {'p': '+', 'm': '-'}[sign] + pos
title = f'{pos_name} {nom_base} to {denom_base} ratio'

for fn in snakemake.output.og_phenograms:
    og_id = os.path.split(fn)[1].split('.')[0]
    pav = og_cnv.loc[og_id] > 0
    plot_phenogram(
        tree,
        pheno.mean(0).to_dict(),
        pav,
        title=title
    )
    plt.savefig(fn)
    plt.show()
```

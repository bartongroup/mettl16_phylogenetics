```python
def rev_comp(seq):
    return seq.translate(RC)[::-1]


def read_intron_seqs(bed_fn, fasta_fn, signal_type):
    introns = set()
    with open(bed_fn) as bed:
        for record in bed:
            chrom, start, end, _, _, strand, *_ = record.strip().split('\t')
            start, end = int(start), int(end)
            introns.add((chrom, start, end, strand))
    introns = list(introns)
    seqs = []
    with pysam.FastaFile(fasta_fn) as fasta:
        for chrom, start, end, strand in introns:
            if signal_type == '5ss':
                left_pos = start - 3 if strand == '+' else end - 7
                right_pos = start + 7 if strand == '+' else end + 3
                seq = fasta.fetch(chrom, left_pos, right_pos).upper().replace('T', 'U')
                if strand == '-':
                    seq = rev_comp(seq)
            elif signal_type == '3ss':
                left_pos = end - 5 if strand == '+' else start - 3
                right_pos = end + 3 if strand == '+' else start + 5
                seq = fasta.fetch(chrom, left_pos, right_pos).upper().replace('T', 'U')
                if strand == '-':
                    seq = rev_comp(seq)
            if seq is not None:
                seqs.append(seq)
    return introns, seqs


def calculate_normalised_counts(seqs):
    seqlen = len(seqs[0])
    counts = np.zeros((seqlen, 4), dtype='f')
    for seq in seqs:
        if len(seq) != seqlen:
            raise ValueError('Not all sequences are of the same length')
        for i, base in enumerate(seq):
            try:
                counts[i] += IUPAC_OHE[base]
            except KeyError:
                continue
    pssm = pd.DataFrame(counts / counts.sum(1)[:, np.newaxis], columns=['A', 'C', 'G', 'U'])
    pssm.index.name = 'position'
    return pssm


def ohe_seq_iupac(seq):
    seq_ohe = []
    for n in seq:
        seq_ohe.append(IUPAC_OHE[n])
    return np.array(seq_ohe)


def pssm_score(seq, pssm, start=None, end=None):
    if start is None:
        start = 0
    if end is None:
        end = len(seq)
    pssm = np.array(pssm)[start: end]
    seq_ohe = ohe_seq_iupac(seq[start: end])
    return (seq_ohe * pssm).ravel().sum()


def species_u5_u6_corr(bed_fn, fasta_fn, max_introns=10_000, bootstraps=10):
    introns, seqs = read_intron_seqs(
        bed_fn, fasta_fn, '5ss'
    )
    nseqs = len(seqs)
    n = min(nseqs, max_introns)
    atou = []
    u56rho = []
    for _ in range(bootstraps):
        n = min(len(seqs), max_introns)
        idx = np.random.randint(0, n, n)
        seqs_sample = [seqs[i] for i in idx]
        pssm = calculate_normalised_counts(seqs_sample)
        atou.append(pssm.loc[6, 'A'] - pssm.loc[6, 'U'])
        u5_score = [pssm_score(s, pssm, 0, 3) for s in seqs_sample]
        u6_score = [pssm_score(s, pssm, 5, 9) for s in seqs_sample]
        r, p = stats.spearmanr(u5_score, u6_score)
        u56rho.append(r)
    return atou, u56rho, nseqs
```

```python
r = Parallel(n_jobs=8)(#snakemake.threads)(
    delayed(species_u5_u6_corr)(bed_fn, fasta_fn)
    for bed_fn, fasta_fn in zip(snakemake.input.juncs, snakemake.input.fastas)
)
species= [os.path.split(fn)[1].split('.')[0] for fn in snakemake.input.juncs]
atou, u56rho, nseqs = zip(*r)
atou = pd.DataFrame(atou, index=species).T
u56rho = pd.DataFrame(u56rho, index=species).T
nseqs = pd.Series(nseqs, index=species)
```

```python
fh, pheno_fn = tempfile.mkstemp(dir=os.environ['TMPDIR'])

u56rho.to_csv(pheno_fn, sep='\t')

res = bootpheno_og_pgls(
    snakemake.input.key_ogs,
    snakemake.input.key_prot_ids,
    pheno_fn,
    snakemake.input.tree,
    njobs=8,#snakemake.threads
)

os.close(fh)
os.remove(pheno_fn)
```

```python
fig = plt.figure(figsize=(14, 10))
ax = plt.subplot2grid(shape=(3, 10), loc=(1, 0), fig=fig, rowspan=2, colspan=8)
hist_ax = plt.subplot2grid(shape=(3, 10), loc=(0, 0), fig=fig, sharex=ax, colspan=8)
hist2_ax = plt.subplot2grid(shape=(3, 10), loc=(1, 8), fig=fig, rowspan=2, colspan=2, sharey=ax)
cbar_ax = plt.subplot2grid(shape=(3, 10), loc=(0, 8), fig=fig, rowspan=1, colspan=1)
cmap = LinearSegmentedColormap.from_list(name='AtoU', colors=[pal[1], '#AAAAAA', pal[0]], N=100)
sm = ScalarMappable(norm=Normalize(-1, 1, clip=True), cmap=cmap)
xvals = np.log10(nseqs)
yvals = u56rho.mean(0)
yerrvals = u56rho.quantile([0.025, 0.975]).values
yerrvals = np.abs(yerrvals - yvals[np.newaxis, :])
cvals = sm.to_rgba(atou.mean(0))
apref = pd.cut(atou.mean(0), [-np.inf, -0.2, 0.2, np.inf], labels=['U', 'W', 'A'])
ax.scatter(xvals, yvals, c=cvals, zorder=3)

for x, y, yerr, c in zip(xvals, yvals, yerrvals.T, cvals):
    if not np.isnan(y):
        ax.errorbar(
            [x], [y], yerr=[[yerr[0]], [yerr[1]]],
            color=c,
            ls='none',
            capsize=2,
            zorder=2,
        )

data_h = []
data_v = []
zorder = 0
hist_h_range = (xvals.min() - 0.1, xvals.max() + 0.1)
hist_v_range = (yvals.min() - 0.1, yvals.max() + 0.1)
for apref, group in nseqs.groupby(apref):
    data_h += xvals[group.index].values.tolist()
    data_v += yvals[group.index].values.tolist()
    hist_ax.hist(
        data_h,
        range=hist_h_range,
        bins=30,
        zorder=zorder,
        color={'A': pal[0], 'U': pal[1], 'W': '#AAAAAA'}[apref],
        label=apref
    )
    hist2_ax.hist(
        data_v,
        range=hist_v_range,
        orientation='horizontal',
        bins=30,
        zorder=zorder,
        color={'A': pal[0], 'U': pal[1], 'W': '#AAAAAA'}[apref],
        label=apref
    )
    zorder -= 1
    sns.regplot(
        x=xvals[group.index],
        y=yvals[group.index],
        color={'A': pal[0], 'U': pal[1], 'W': '#AAAAAA'}[apref],
        scatter=False,
        truncate=False,
        line_kws={'zorder': 1},
        ax=ax
    )

for sp in snakemake.config['funannotate']['trusted_annotations']:
    ax.annotate(
        xy=(xvals[sp], yvals[sp]),
        xytext=(xvals[sp] + (np.random.random() - 0.25),
                yvals[sp] + (np.random.random() - 0.25)),
        s=f'{sp[0]}. {sp.split("_")[1]}',
        fontsize=12,
        arrowprops=dict(facecolor='black', width=2, edgecolor='none')
    )

hist_ax.legend(title='Overall 5\'SS+4\npreference')
hist_ax.set_ylabel('Species')
hist2_ax.set_xlabel('Species')

plt.colorbar(sm, label='+4 A to U ratio', cax=cbar_ax)
ax.set_ylabel('Spearman\'s rho of U5 and U6 snRNA \ninteracting position PSSM scores (U5/6ρ)')
ax.set_xlabel('Number of identified 5\'SSs (log10 scale)')
ax.set_xticks([2, 3, 4])
ax.set_xlim(*hist_h_range)
ax.set_xticklabels([100, 1000, 10_000])
plt.tight_layout()
plt.savefig(snakemake.output.scatter)
plt.show()
```

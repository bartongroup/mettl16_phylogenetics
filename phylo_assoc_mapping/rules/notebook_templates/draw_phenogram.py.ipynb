```python
import os
import re
from glob import glob
import tempfile
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap, Normalize, to_hex
from matplotlib.cm import ScalarMappable
import seaborn as sns

import ete3
os.environ['QT_QPA_PLATFORM']='offscreen'

from rpy2 import robjects as robj
from rpy2.robjects.packages import importr, isinstalled
from rpy2.robjects.conversion import localconverter
from rpy2.rinterface import RRuntimeWarning
from rpy2.rinterface_lib.embedded import RRuntimeError

from rpy2.robjects import pandas2ri

warnings.filterwarnings("ignore", category=RRuntimeWarning)


def import_or_install(package_name, repo='https://cran.ma.imperial.ac.uk/'):
    if not isinstalled(package_name):
        utils = importr('utils')
        utils.install_packages(robj.StrVector([package_name]), repos=repo)
    return importr(package_name)


ape = import_or_install('ape')

## Default plotting params
%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
IUPAC = {
    'A': ['A'],
    'C': ['C'],
    'G': ['G'],
    'U': ['U'],
    'R': ['A', 'G'],
    'Y': ['C', 'U'],
    'S': ['G', 'C'],
    'W': ['A', 'U'],
    'K': ['G', 'U'],
    'M': ['A', 'C'],
    'B': ['C', 'G', 'U'],
    'D': ['A', 'G', 'U'],
    'H': ['A', 'C', 'U'],
    'V': ['A', 'C', 'G'],
    'N': ['A', 'C', 'G', 'U']
}

pssm_dir = os.path.split(snakemake.input.pssms[0])[0]
pssm_fns = glob(f'{pssm_dir}/*.{snakemake.wildcards.signal_type}.tsv')

pssms = {
    os.path.split(fn)[1].split('.')[0]: pd.read_csv(fn, sep='\t', index_col=[0, 1])
    for fn in pssm_fns
}

bar_pheno = {}
pheno = {}
for node, pssm in pssms.items():
    mean_pssm = pssm.loc[pd.IndexSlice[:, abspos], :].mean(0)
    # sankoff parsimony phenos used for stacked bar charts
    bar_pheno[node] = mean_pssm
    if node in species:
        # internal phenotypes for phenograms will be calculated with APE so only take pheno for leaves
        pheno[node] = mean_pssm[IUPAC[nom_base]].sum() - mean_pssm[IUPAC[denom_base]].sum()
```

```python
def to_ape_tree(tree):
    tmp_fh, tmp_fn = tempfile.mkstemp()
    tree.write(outfile=tmp_fn, format=1, format_root_node=True)
    r_tree = ape.read_tree(tmp_fn)
    os.close(tmp_fh)
    os.remove(tmp_fn)
    return r_tree


def get_ancestral_states(tree, pheno):
    r_tree = to_ape_tree(tree)
    leaf_order = list(r_tree.rx2['tip.label'])
    pheno = [pheno[n] for n in leaf_order]
    anc = ape.ace(
        robj.FloatVector(pheno),
        r_tree,
        type='continuous'
    )
    return dict(zip(r_tree.rx2['node.label'], anc.rx2['ace']))


def _plot_recursive(node, pheno, x1, y1, colour, ax, clade_root_node_colours, label_species):
    for child in node.children:
        x2 = x1 + child.dist
        y2 = pheno[child.name]
        node_leaves = set([n.name for n in child.get_leaves()])
        if node_leaves.intersection(label_species):
            lw = 3
            zorder = 1
            alpha = 1
        else:
            lw = 0.8
            zorder = 0
            alpha = 0.75
        ax.plot([x1, x2], [y1, y2], color=colour, lw=lw, zorder=zorder, alpha=alpha)
        if not child.is_leaf():
            if child.name in clade_root_node_colours:
                _plot_recursive(child, pheno, x2, y2,
                                clade_root_node_colours[child.name],
                                ax, clade_root_node_colours,
                                label_species)
            else:
                _plot_recursive(child, pheno, x2, y2, colour, ax,
                                clade_root_node_colours, label_species)
        else:
            if child.name in label_species:
                ax.annotate(xy=(x2, y2), s=child.name,
                            ha='left', va='center', color=colour)

    
def plot_stacked_pheno_histogram(pheno, species_clades, clade_colours, ax):
    hist_data = []
    colours = []
    labels = []
    for clade_name, species_list in species_clades.items():
        hist_data.append([pheno[sp] for sp in species_list])
        colours.append(clade_colours[clade_name])
        labels.append(clade_name)

    for i in range(len(hist_data)):
        ax.hist(
            np.concatenate(hist_data[i:]),
            orientation='horizontal',
            range=(-1, 1),
            bins=20,
            color=colours[i],
            label=labels[i],
            edgecolor='none',
        )
    ax.legend(title='Clades')
    return ax


def plot_phenogram(tree, pheno,
                   species_clades, clade_colours,
                   title=None, ultrametric=True,
                   label_species=None, figsize=(15, 10), add_hist=True):
    tree = tree.copy()
    if ultrametric:
        tree.convert_to_ultrametric(tree_length=1)
    anc_states = get_ancestral_states(tree, pheno)
    anc_states.update(pheno)

    clade_root_node_colours = {}

    for clade_name, species_list in species_clades.items():
        clade_root_node_colours[
            tree.get_common_ancestor(species_list).name
        ] = clade_colours[clade_name]

    fig = plt.figure(figsize=figsize)
    if add_hist:
        phenogram_ax = plt.subplot2grid((1, 3), (0, 0), colspan=2)
        hist_ax = plt.subplot2grid((1, 3), (0, 2), colspan=1, sharey=phenogram_ax)
        plot_stacked_pheno_histogram(pheno, species_clades, clade_colours, hist_ax)
        hist_ax.set_xlabel('Species')
    else:
        phenogram_ax = plt.subplots()
        hist_ax = None
    if label_species is None:
        label_species = []
    x1 = 0
    y2 = anc_states[tree.name]
    _plot_recursive(
        tree, anc_states, x1, y2,
        clade_root_node_colours[tree.name],
        phenogram_ax,
        clade_root_node_colours,
        label_species)
    if ultrametric:
        phenogram_ax.set_xlabel('Branch length (ultrametric)')
    else:
        phenogram_ax.set_xlabel('Branch length')
    if title is not None:
        phenogram_ax.set_ylabel(title)
    plt.tight_layout()
    return fig, phenogram_ax, hist_ax
```

```python
IUPAC_COLORS = [pal.as_hex()[2], pal.as_hex()[0], pal.as_hex()[3], pal.as_hex()[1]]

def draw_collapsed_tree_with_bars(tree, bar_data, species_clades, reference_species, clade_colours):
    
    t = tree.copy()

    keep_nodes = set(reference_species)
    for sp in reference_species:
        bd = bar_data[sp]
        bd = bd.values / bd.sum() * 100
        barface = ete3.StackedBarFace(bd, 150, 15, colors=IUPAC_COLORS)
        n = t.search_nodes(name=sp)[0]
        n.add_face(barface, 0, 'aligned')
        try:
            n.img_style['bgcolor'] = clade_colours[species_clades[sp]]
        except KeyError:
            n.img_style['bgcolor'] = clade_colours['Outgroups']
    for clade_name, species_list in species_clades.items():
        lca = t.get_common_ancestor(species_list)
        if lca.name != t.get_tree_root().name:
            bd = bar_data[lca.name]
            bd = bd.values / bd.sum() * 100
            lca.name = clade_name
            keep_nodes.add(clade_name)
            keep_nodes.update([n.name for n in lca.get_ancestors()])
            barface = ete3.StackedBarFace(bd, 150, 15, colors=IUPAC_COLORS)
            lca.add_face(barface, 0, 'aligned')
            lca.img_style['bgcolor'] = clade_colours[clade_name]

    t.prune(keep_nodes)
    t.convert_to_ultrametric()
    t.ladderize()

    cmap = LinearSegmentedColormap.from_list(name='redblack', colors=['#ff0000', '#252525'], N=100)
    # colour branches by confidence
    for node in t.traverse():
        color = to_hex(cmap(float(node.conf)))
        node.img_style['vt_line_color'] = color

    ts = ete3.TreeStyle()
    ts.show_leaf_name = True

    return t, ts, cmap
    
```

```python
def read_orthofinder_n0(n0_fn):
    og = pd.read_csv(n0_fn, sep='\t', index_col='HOG').fillna('')
    og = og.iloc[:, 2:]
    return og

def get_key_ogs(key_prot_ids, og):
    og_ids = []
    for _, gene_symbol, prot_id, species_name in key_prot_ids.itertuples():
        try:
            og_id = og.query(f'{species_name}.str.contains("{prot_id}")').index[0]
        except IndexError:
            warnings.warn(f'Could not find orthogroup for "{gene_symbol}"')
            og_id = np.nan
        og_ids.append(og_id)
    key_prot_ids['og_id'] = og_ids
    og_idx = []
    new_og_ids = []
    for og_id, group in key_prot_ids.groupby('og_id'):
        og_idx.append(og_id)
        new_og_ids.append('_'.join(group.gene_symbol))
    og_key = og.loc[og_idx]
    og_key.index = new_og_ids
    return og_key


key_og_n0 = read_orthofinder_n0(snakemake.input.key_ogs)
key_prot_table = pd.read_csv(
    snakemake.input.key_prot_ids, sep='\t', names=['gene_symbol', 'prot_id', 'species']
)
key_og_n0 = get_key_ogs(key_prot_table, key_og_n0)
```

```python
def draw_full_circular_tree_with_bars(tree, bar_data, og_data, species_clades, clade_colours):
    
    t = tree.copy()

    for node in t.get_leaves():
        bd = bar_data[node.name]
        bd = bd.values / bd.sum() * 100
        for clade_name, sp_list in species_clades.items():
            if node.name in sp_list:
                colour = clade_colours[clade_name]
                break
        else:
            colour = clade_colours['Outgroups']
        genus, species = node.name.split('_', 1)
        name = f'{genus[0]}. {species}'
        label = ete3.RectFace(300, 25,
                              label={'text': name, 'color': '#252525', 'fontsize':20, 'font': 'Arial'},
                              bgcolor=colour, fgcolor=None)
        node.add_face(label, 0, 'aligned')
        # spacer
        node.add_face(ete3.RectFace(10, 25, bgcolor=None, fgcolor=None), 1, 'aligned')
        barface = ete3.StackedBarFace(bd, 300, 25, colors=IUPAC_COLORS)
        barface.rotable = False
        node.add_face(barface, 2, 'aligned')
        i = 3
        for og_name, ogd in og_data.items():
            node.add_face(ete3.RectFace(10, 25, bgcolor=None, fgcolor=None), i, 'aligned')
            i += 1
            o = ete3.CircleFace(15, color='#252525' if ogd[node.name] else '#aaaaaa')
            node.add_face(o, i, 'aligned')
            i += 1

    t.ladderize()

    cmap = LinearSegmentedColormap.from_list(name='redblack', colors=['#ff0001', '#252526'], N=100)
    # colour branches by confidence
    for node in t.traverse():
        color = to_hex(cmap(float(node.conf)))
        node.img_style['vt_line_color'] = color
        node.img_style['vt_line_width'] = 5
        node.img_style['hz_line_width'] = 5
        node.img_style["size"]=0

    ts = ete3.TreeStyle()
    ts.show_leaf_name = False
    ts.mode = 'c'
    ts.arc_start = 5
    ts.arc_span = 350

    return t, ts, cmap
```

```python
og_data = {
    'METTL16': key_og_n0.loc['METTL16'].to_dict(),
    'SNRNP27': key_og_n0.loc['SNRNP27'].to_dict(),
}

t, ts, cmap = draw_full_circular_tree_with_bars(
    tree, bar_pheno, og_data,
    snakemake.config['known_clades'],
    clade_colours
)
t.render(snakemake.output.full_circular_tree, tree_style=ts);
t.render("%%inline", tree_style=ts)
```
